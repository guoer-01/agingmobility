# spatial distribution of high-aging CBGs

```python
import sys
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import MultiPolygon, Polygon
from shapely.ops import  unary_union
import math
import pandas as pd
import plot_utils as pu
from matplotlib.colors import LinearSegmentedColormap,ListedColormap
import matplotlib.cm as cm
```

```python
#cbg-level geodata
us_geo=gpd.read_file('D:/main/1_research/P10_Mobility/p10_aging mobility/new data/census/safegraph_open_census_data_2010_to_2019_geometry/cbg.geojson')
```

```python
us_geo['countycode']=us_geo.apply(lambda x: x['StateFIPS']+x['CountyFIPS'],axis=1)
us_geo=us_geo.rename(columns={'CensusBlockGroup':'census_block_group'})
us_geo['census_block_group']=us_geo['census_block_group'].astype('int64')
```

```python
#county-level geodata

gpdCountyUS=gpd.GeoDataFrame(columns=['county','geometry'])
for i,county in enumerate(us_geo['countycode'].unique()):
    polist=[]
    for multipol in us_geo[us_geo['countycode']==county]['geometry'].to_list():
        for poligon in multipol:
            polist.append(poligon) 
    gpdCountyUS.loc[i]=[county,unary_union(polist)]
    print(county)
    
#delete the abnormal county 02016 in AK 15003 in HI
#gpdCountyUS=gpdCountyUS.drop( gpdCountyUS[gpdCountyUS['county']=='02016'].index )
#gpdCountyUS=gpdCountyUS.drop( gpdCountyUS[gpdCountyUS['county']=='15003'].index )

dict_state={'1': 'AL', '2': 'AK', '4': 'AZ', '5': 'AR', '6': 'CA', '8': 'CO', '9': 'CT', '10': 'DE', '11': 'DC', '12': 'FL', '13': 'GA', '15': 'HI', '16': 'ID',
            '17': 'IL', '18': 'IN', '19': 'IA', '20': 'KS', '21': 'KY', '22': 'LA', '23': 'ME', '24': 'MD', '25': 'MA', '26': 'MI', '27': 'MN', '28': 'MS', '29': 'MO',
            '30': 'MT', '31': 'NE', '32': 'NV', '33': 'NH', '34': 'NJ', '35': 'NM', '36': 'NY', '37': 'NC', '38': 'ND', '39': 'OH', '40': 'OK', '41': 'OR', '42': 'PA', 
            '44': 'RI', '45': 'SC', '46': 'SD', '47': 'TN', '48': 'TX', '49': 'UT', '50': 'VT', '51': 'VA', '53': 'WA', '54': 'WV', '55': 'WI', '56': 'WY', '60': 'AS', 
            '66': 'GU', '69': 'MP', '72': 'PR', '74': 'UM', '78': 'VI'}



```

```python
df_census=pd.read_csv('D:/main/1_research/P10_Mobility/p10_aging mobility/new data/census/all_census_data_with_location.csv',
                     usecols=['census_block_group','population_by_age','percent_above_65','location'])

df_census['is_old']=df_census.apply(lambda x: 1 if x['percent_above_65']>0.5 else 0,axis=1)
df_census['census_block_group']=df_census['census_block_group'].astype('str')
df_census['county']=df_census.apply(lambda x: '0'+x['census_block_group'][0:4] if len(x['census_block_group'])==11 else
                                  x['census_block_group'][0:5],axis=1 )
```

```python
state_color_dict={'US':'#98A3CA','FL':'#D0DCE8', 'AZ':'#E6BCB0','CA':'#F2DB96'}
color_dict={'Health Care Services':'#7FB2D5','Grocery and Food Supply':'#AF9BC9', 'Housing and Real Estate':'#A9C7C4'}

colors=['#B3D8D5', '#EDA1A4']
cmap=ListedColormap(colors)
county_data2=gpdCountyUS.set_index('county')['isold'].to_dict()

fig=pu.choropleth_map_county(
    county_data2,
    map_title='US',
    vmax=1, dpi=300, fontsize=14,cmap=cmap
)
#plt.savefig('D:/main/1_research/P10_Mobility/p10_aging mobility/colorselect/{}.png'.format('new_US_colorTry3'),dpi=300)
plt.savefig('D:/main/1_research/P10_Mobility/p10_aging mobility/new data/census/{}.png'.format('new_US'),dpi=600)
plt.show()
```

```python
gpdCountyUS['isold']=gpdCountyUS.apply(lambda x: sum(df_census[df_census['county']==x['county']]['is_old'])
                                       if x['county'] in df_census['county'].unique() else 0, axis=1)
gpdCountyUS['isold']=gpdCountyUS.apply(lambda x: 1 if x['isold']>0 else 0, axis=1)
```

```python
gpdCountyUS['state']=gpdCountyUS.apply(lambda x: dict_state[str(int(x['county'][0:2]))],axis=1)
gpdCountyUS['age']=gpdCountyUS.apply(lambda x: 'greater than 0.5' if x['isold']==1 else 'less than or equal to 0.5',axis=1)

#color 
gpdCountyUS['color']=gpdCountyUS.apply(lambda x: '#fc8d62' if x['isold']==1 else '#b3e2cd',axis=1)
```

```python
#state-level geodata
from shapely.geometry import MultiPolygon, Polygon
from shapely.ops import  unary_union
gpdUS=gpd.GeoDataFrame(columns=['State','geometry'])
for i,state in enumerate(gpdCountyUS['state'].unique()):

    polist=[]

    for multipol in gpdCountyUS[gpdCountyUS['state']==state]['geometry'].to_list(): 
        if type(multipol)==Polygon:
            polist.append(multipol)
        else:
            for poligon in multipol:
                polist.append(poligon)


    gpdUS.loc[i]=[state,unary_union(polist)]
    print(state)

```

```python
df_census['census_block_group']=df_census['census_block_group'].astype('int64')
dff=pd.merge(us_geo,df_census)

dff['age']=dff.apply(lambda x: 'greater than 0.5' if x['is_old']==1 else 'less than or equal to 0.5',axis=1)
#colors={'less than or equal to 0.5':'#b3e2cd','greater than 0.5':'#fc8d62’}

dff['color']=dff.apply(lambda x: '#fc8d62' if x['is_old']==1 else '#b3e2cd',axis=1)
```
